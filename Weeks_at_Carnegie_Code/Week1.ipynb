```python
from astropy.table import Table
t = Table.read('snapshot_now.fits') # path_to_file is the path where you downloaded the attached file
```

```python
t
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
# two figures, side by side:
fig, axarr = plt.subplots(1,2,figsize=(10,5))
# axarr is an array of axis objects, with each element representing one subplot

# first subplot
ax=axarr[0]
ax.plot(t['x'],t['y'],color='blue',marker='o',ms=1,ls='None', alpha = 0.01)
ax.set_xlabel('X (kiloparsec)', fontsize=20)
ax.set_ylabel('Y (kiloparsec)', fontsize=20)

# second subplot
ax=axarr[1]
ax.plot(t['x'],t['z'],color='red',marker='o',ms=1,ls='None', alpha = 0.01)
ax.set_xlabel('X (kiloparsec)', fontsize=20)
ax.set_ylabel('Z (kiloparsec)', fontsize=20)


# change title on subplots
#ax=axarr[0]
#ax.set_xlabel('x vs y',fontsize=16)
#ax=axarr[1]
#ax.set_xlabel('x vs z',fontsize=16)
fig.tight_layout() #this adjusts the size of the graphs 
plt.savefig('Week1 Figures.png')
```
